```python
# Another solution


for cat in list_categories:
    categories[cat] = (categories[cat] - min_val_by_cat.min()) / (max_val_by_cat.max() - min_val_by_cat.min())


categories
```

```python

from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

from yellowbrick.cluster import KElbowVisualizer

# Generate synthetic dataset with 8 random clusters
y = make_blobs(n_samples=1000, n_features=12, centers=8, random_state=42)

# Instantiate the clustering model and visualizer
model = KMeans()
visualizer = KElbowVisualizer(model, k=(4,12))

visualizer.fit(categories)        # Fit the data to the visualizer
visualizer.show()        # Finalize and render the figure
```

```python
cs = []
for i in range(1, 15):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(categories)
    cs.append(kmeans.inertia_)
plt.plot(range(1, 15), cs)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('CS')
plt.show()
```

```python

# found with elbow after executing the algorithm
CLUSTER_NUM = 6

mm_model = KMeans(n_clusters=CLUSTER_NUM, random_state=0)
mm_model.fit(categories_regions)
```

```python


cluster_assignments = mm_model.labels_
categories['cluster'] = cluster_assignments

# correct_labels = sum(y == cluster_assignments)
    
categories[categories['cluster'] == 0]
```

```python
from sklearn.decomposition import PCA 
import seaborn as sns


pca = PCA(n_components=2)
s_d_r = pca.fit_transform(categories)

sns.pairplot(categories)
```

```python
# Calculate the average amount by cluster 

# the range 0 ... (CLUSTER_NUM - 1)

averages = []

# for i in range(0, CLUSTER_NUM - 1): 
    # rows_count = categories[categories["cluster"] == i].size
    # averages.append(categories[categories["cluster"] == i].sum() / rows_count)

for i in range(0, CLUSTER_NUM): 
    dd = categories[categories["cluster"] == i]
    dd.drop(["cluster"], axis=1, inplace=True)

    # dd
    averages.append((dd.sum().sum() / (dd.size / 10)) * max_val_by_cat.max())



plt.plot(averages)
plt.show()

averages
```

```python
# Highest paid category by cluster

pop_categories = []

# for i in range(0, CLUSTER_NUM - 1):
# rows_count = categories[categories["cluster"] == i].size
# averages.append(categories[categories["cluster"] == i].sum() / rows_count)

for i in range(0, CLUSTER_NUM):
    dd = categories[categories["cluster"] == i]
    dd.drop(["cluster"], axis=1, inplace=True)

    # dd
    gg = dd.sum().to_dict()

    max_categ = max(gg, key=gg.get)
    pop_categories.append(max_categ)

pop_categories
plt.plot(pop_categories)
plt.show()

# averages
```
